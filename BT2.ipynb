```python
import pandas as pd
import numpy as np
url = 'http://winterolympicsmedals.com/medals.csv'
data = pd.read_csv(url)
data_FRA = data [data['NOC'] == 'FRA']
# Hiển thị danh sách dữ liệu
data_FRA.head(83)
```

```python
# Tổng số huy chương từ năm 1924-2006
data_FRA['Medal'].count()
```

```python
# Tổng số huy chương từng năm
data_FRA.groupby('Year').count().drop(['City','Sport','Discipline','NOC','Event','Event gender'], axis =1).head(83)
```

```python
# Năm có huy chương nhiều nhất
max = data_FRA.groupby('Year')['Medal'].count().max()
data_FRA_Year = data_FRA.groupby('Year').count()
data_FRA_Year[data_FRA_Year['Medal'] == max].drop(['City','Sport','Discipline','NOC','Event','Event gender'], axis =1).head(83)
```

```python
# Năm có huy chương nhỏ nhất
min = data_FRA.groupby('Year')['Medal'].count().min()
data_FRA_Year = data_FRA.groupby('Year').count()
data_FRA_Year[data_FRA_Year['Medal'] == min].drop(['City','Sport','Discipline','NOC','Event','Event gender'], axis =1).head(83)
```

```python
#Tổng số huy chương Đồng, Bạc, Vàng
print('Tổng số huy chương Đồng, Vàng, Bạc:')
data_FRA.groupby('Medal')['Medal'].count()
```

```python
# Các loại huy chương qua từng năm
print('Các loại huy chương qua từng năm:')
data_FRA.groupby(['Year','Medal'])['Medal'].count()
```

```python
#Năm có huy chương vàng nhiều nhất
data_FRA_Gold = data_FRA[data_FRA['Medal'] == 'Gold']
max = data_FRA_Gold.groupby('Year')['Medal'].count().max()
dataframe = data_FRA_Gold.groupby('Year').count()
dataframe[dataframe['Medal'] == max].drop(['City','Sport','Discipline','NOC','Event','Event gender'], axis =1).head(83)
```
