```python
#Housing
import pandas as pd
import matplotlib.pyplot as plt
url = 'Housing.csv'
dataHousing = pd.read_csv(url)
dataHousing.hist()
plt.show
```

```python
correlationHousing = dataHousing.corr()
print(correlationHousing)
```

```python
import seaborn as sns
sns.heatmap(correlationHousing, annot = True)
```

```python
sns.pairplot(dataHousing)
```

```python
# Price phụ thuộc vào Area lớn nhất 0.54
dataHousing.plot(kind = 'scatter', x = 'area', y = 'price', figsize=(9,9), color = 'black')
```

```python
#Mô hình bậc 1
from sklearn import linear_model
modelHousing1 = linear_model.LinearRegression()
X = pd.DataFrame(dataHousing['area'])
modelHousing1.fit(X, y = dataHousing['price'])
print('Intercept của mô hình: ', modelHousing1.intercept_)
print('Coeffient của mô hình: ', modelHousing1.coef_)
print('Độ chính xác của mô hình: ', modelHousing1.score(X,y = dataHousing['price']))
dataHousing.plot(kind = 'scatter', x = 'area', y = 'price', figsize=(9,9), color = 'black')
y_pred = modelHousing1.predict(X)
plt.plot(X,y_pred, color = 'blue')
```

```python
#Mô hình bậc 2
from sklearn import linear_model
modelHousing3 = linear_model.LinearRegression()
X = pd.DataFrame([dataHousing['area'], dataHousing['area']**2]).T
modelHousing3.fit(X, y = dataHousing['price'])
print('Intercept của mô hình: ', modelHousing3.intercept_)
print('Coeffient của mô hình: ', modelHousing3.coef_)
print('Độ chính xác của mô hình: ', 
      modelHousing3.score(X,y = dataHousing['price']))

import numpy as np
range = np.arange(2000,16000,1)
dataHousing.plot(kind = 'scatter', x = 'area', y = 'price', figsize=(9,9),
                 color = 'black', xlim=(2000,16000))
predictor = pd.DataFrame([range, range**2]).T
y_pred = modelHousing3.predict(X = predictor)
plt.plot(range,y_pred, color = 'blue')
```

```python
#Mô hình bậc 5
from sklearn import linear_model
modelHousing5 = linear_model.LinearRegression()
X = pd.DataFrame([dataHousing['area'], dataHousing['area']**2, 
                  dataHousing['area']**3, dataHousing['area']**4,
                  dataHousing['area']**5]).T
modelHousing5.fit(X, y = dataHousing['price'])
print('Intercept của mô hình: ', modelHousing5.intercept_)
print('Coeffient của mô hình: ', modelHousing5.coef_)
print('Độ chính xác của mô hình: ', 
      modelHousing5.score(X,y = dataHousing['price']))

import numpy as np
range = np.arange(2000,16000,1)
dataHousing.plot(kind = 'scatter', x = 'area', y = 'price', figsize=(9,9),
                 color = 'black', xlim=(2000,16000))
predictor = pd.DataFrame([range, range**2, range**3, range**4, range**5]).T
y_pred = modelHousing5.predict(X = predictor)
plt.plot(range,y_pred, color = 'blue')
```

```python
#Mô hình bậc 10
from sklearn import linear_model
modelHousing10 = linear_model.LinearRegression()
X = pd.DataFrame([dataHousing['area'], dataHousing['area']**2, 
                  dataHousing['area']**3, dataHousing['area']**4,
                  dataHousing['area']**5, dataHousing['area']**6,
                  dataHousing['area']**7, dataHousing['area']**8,
                  dataHousing['area']**9, dataHousing['area']**10]).T
modelHousing10.fit(X, y = dataHousing['price'])
print('Intercept của mô hình: ', modelHousing10.intercept_)
print('Coeffient của mô hình: ', modelHousing10.coef_)
print('Độ chính xác của mô hình: ', 
      modelHousing10.score(X,y = dataHousing['price']))

import numpy as np
range = np.arange(2000,16000,1)
dataHousing.plot(kind = 'scatter', x = 'area', y = 'price', figsize=(9,9),
                 color = 'black', xlim=(2000,16000))
predictor = pd.DataFrame([range, range**2, range**3, range**4, range**5,
                          range**6, range**7, range**8, range**9, range**10]).T
y_pred = modelHousing10.predict(X = predictor)
plt.plot(range,y_pred, color = 'blue')
```
