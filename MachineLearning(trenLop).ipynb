```python
#hồi quy và phân loại
#có giám sát, không giám sát, tăng cường
import pandas as pd #thư viện xử lý đồ họa
url = 'mtcars.csv'
data = pd.read_csv(url) 
import matplotlib.pyplot as plt
data.hist()
plt.show
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x ='wt',y='mpg', figsize=(9,9), color = 'black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['wt']), y = data['mpg']) # ham cho may hoc
print(model.intercept_)
print(model.coef_)
# y = coef*x + intercept
```

```python
#do chinh xac accurary
X = pd.DataFrame(data['wt'])
y = data['mpg']
model.score(X,y)
```

```python
data.plot(kind='scatter', x='wt', y='mpg', figsize=(9,9), color='black',xlim=(0,7)) # dau cham den
y_pred =model.predict(X = pd.DataFrame(data['wt'])) # ham bac 1 dai dien cho du lieu
plt.plot(data['wt'], y_pred, color='blue')
```

```python
#bac 2
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2]).T # .T la chuyen vij hang thanh cot, cot thanh hang
model.fit(X,y=data['mpg'])
print('Intercept cua mo hinh:', model.intercept_)
print('Coeffient cua mo hinh:', model.coef_)
print('Do chuan xac cua mo hinh:', model.score(X,y=data['mpg']))
```

```python
import numpy as np
range = np.arange(1.5,5.5,0.1)

data.plot(kind='scatter', x='wt', y='mpg', figsize=(9,9), color='black',xlim=(0,7)) #du lieu cham black

predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color='blue')
```

```python
# Hoi quy 2 bien wt va hp
model = linear_model.LinearRegression()
model.fit(X = data.loc[:,['wt','hp']], y = data['mpg'])
print(model.intercept_)
print(model.coef_)
print(model.score(X=data.loc[:,['wt','hp']], y = data['mpg']))
# y = -3.877wt - 0.031hp+ 37,22
```
